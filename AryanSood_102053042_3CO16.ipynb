```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/Creditcard_data.csv')
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.ensemble import RandomForestClassifier
def RandomForest(X_train, y_train, X_test, y_test):
  clf = RandomForestClassifier(n_estimators = 2)
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  return accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
def KNN(X_train, y_train, X_test, y_test):
  knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)
  knn_pred = knn.predict(X_test)
  return accuracy_score(y_test,knn_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
def DecisionTree(X_train, y_train, X_test, y_test):
  tree = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train)
  tree_pred = tree.predict(X_test)
  return accuracy_score(y_test,tree_pred)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
def naive_bayes(X_train, y_train, X_test, y_test):
  model=GaussianNB()
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  return accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score

def ExtraT(X_train, y_train, X_test, y_test):
 # Create a RidgeClassifier object and fit it to the training data
 et_clf = ExtraTreesClassifier(n_estimators=40, random_state=1)

 # Train the classifier on the training set
 et_clf.fit(X_train, y_train)
 # Make predictions on the testing set and calculate accuracy
 y_pred = et_clf.predict(X_test)
 acc = accuracy_score(y_test, y_pred)
 return acc
```

```python
from sklearn.preprocessing import normalize
Amount = normalize([df['Amount']])[0]
df['Amount'] = Amount
df = df.iloc[:, 1:]
```

```python
df.head()
```

```python
clas = df["Class"].value_counts(normalize=False)
print(clas)
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
# over sampling the minority data -> balancing the imbalanced dataset
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_r, y_r = ros.fit_resample(X, y)
```

```python
clas = y_r.value_counts(normalize=False)
print(clas)
```

```python
from sklearn.model_selection import train_test_split
X_dummy, X_test, y_dummy, y_test = train_test_split(X_r, y_r, test_size=0.33, random_state=42)
```

```python
df_new = pd.concat([X_r, y_r], axis=1)
df_new
```

```python
z = 1.96 #95% confidence level
p = 0.5
e = 0.05 #5% esimated margin of error
```

##SIMPLE RANDOM SAMPLING

```python
#Simple Random Sampling 
import math
n_srs = (z*z*p*(1-p))/(e*e)
n_srs = math.floor(n_srs)
print(n_srs)
```

```python
# from pandas.core.dtypes.generic import DataFrame
import random
def random_sampling(df, n_srs):
    random_sample = df.sample(n = n_srs)
    return(random_sample)
randomSample = random_sampling(df_new, n_srs)
```

```python
randomSample.head()
```

```python
X_srs = randomSample.drop('Class', axis=1) #X_train
y_srs = randomSample['Class'] #y_train
```

```python
len(y_srs)
```

```python
srs_RandomForest = RandomForest(X_srs,y_srs,X_test,y_test)
srs_DecisionTree = DecisionTree(X_srs,y_srs,X_test,y_test)
srs_KNN = KNN(X_srs,y_srs,X_test,y_test)
srs_naive=naive_bayes(X_srs,y_srs,X_test,y_test)
srs_et=ExtraT(X_srs,y_srs,X_test,y_test)
a1 = [srs_RandomForest,srs_DecisionTree,srs_KNN,srs_et,srs_naive]
```

```python
a1
```

##STRATIFIED SAMPLING

```python
#Stratified Sampling
import math
s = 2 #no of strata
n_strat = (z*z*p*(1-p))/(e*e/s*s)
n_strat = math.floor(n_srs)
print(n_strat)
```

```python
# stratified sampling
from sklearn.model_selection import train_test_split
strata = df_new.groupby('Class')
stratified_df = strata.apply(lambda x: x.sample(n_strat))
stratified_df
```

```python
stratified_df.head()
```

```python
X_st = stratified_df.drop('Class', axis=1) #X_train
y_st = stratified_df['Class'] #y_train

st_RandomForest = RandomForest(X_st,y_st,X_test,y_test)
st_DecisionTree = DecisionTree(X_st,y_st,X_test,y_test)
st_KNN = KNN(X_st,y_st,X_test,y_test)
st_ET = ExtraT(X_st,y_st,X_test,y_test)
st_naive=naive_bayes(X_st,y_st,X_test,y_test)

a2 = [st_RandomForest,st_DecisionTree,st_KNN,st_ET,st_naive]
```

##CLUSTER SAMPLING

```python
from sklearn.cluster import KMeans
#no of clusters = 15
kmeans = KMeans(n_clusters=15, random_state=42).fit(df_new)
cluster_assignments = kmeans.labels_
cluster = [1,3,5,6,8]
cluster_series = pd.Series(cluster_assignments)
cluster_df = df_new[cluster_series.isin(cluster)]
```

```python
cluster_df.head()
```

```python
X_C = cluster_df.drop('Class', axis=1) #X_train
y_C = cluster_df['Class'] #y_train

C_RandomForest = RandomForest(X_C,y_C,X_test,y_test)
C_DecisionTree = DecisionTree(X_C,y_C,X_test,y_test)
C_KNN = KNN(X_C,y_C,X_test,y_test)
C_et = ExtraT(X_C,y_C,X_test,y_test)
C_naive = naive_bayes(X_C,y_C,X_test,y_test)

a3 = [C_RandomForest,C_DecisionTree,C_KNN,C_et,C_naive]
```
